```python
# используем Z-критерий
from math import sqrt

p1 = 0.0105 #conv1

p2 = 0.0147 #conv2

n1 = 164 

n2 = 228

N1 = 15550

N2 = 15550

p =  (n1 * p1 + n2 * p2) / (n1 + n2)

Z = (p1 - p2) / sqrt(p * (1 - p) * (1/N1 + 1/N2))

Z
```
